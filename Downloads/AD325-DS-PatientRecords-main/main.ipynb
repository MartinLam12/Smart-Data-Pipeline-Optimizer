```python
from patient_record_mgmt import PatientRecordManagementSystem
```

```python
# Create the Patient Record Management System and load data from CSV
prms = PatientRecordManagementSystem()
prms.build_tree_from_csv("data/patient_records.csv")
```

```python
# Display all patient records
print("All Patient Records:")
print("-" * 100)
prms.display_all_records()
```

```python
# Visualize the initial tree
dot = prms.visualize_tree()
dot.render("initial_tree", format="png", cleanup=True)
print("Initial tree saved as 'initial_tree.png'")
dot
```

```python
# Search for patient records with IDs 2, 25, and 47
print("Search for patient records with IDs 2, 25, and 47:")
print("-" * 100)

for patient_id in [2, 25, 47]:
    result = prms.search_patient_record(patient_id)
    if result:
        print(f"Patient ID: {result.patient_id}, Name: {result.name}, Age: {result.age}, "
              f"Diagnosis: {result.diagnosis}, Blood Pressure: {result.blood_pressure}, "
              f"Pulse: {result.pulse}, Body Temperature: {result.body_temperature}")
    else:
        print(f"Patient ID {patient_id} not found.")
```

```python
# Delete patient records with IDs 3, 10, and 30
print("Deleting patient records with IDs 3, 10, and 30...")
print("-" * 100)

prms.delete_patient_record(3)
prms.delete_patient_record(10)
prms.delete_patient_record(30)

print("Records deleted successfully!")
print("\nAll Patient Records after deletion:")
print("-" * 100)
prms.display_all_records()
```

```python
# Visualize the tree after deletion
dot = prms.visualize_tree()
dot.render("after-deletion_tree", format="png", cleanup=True)
print("Tree after deletion saved as 'after-deletion_tree.png'")
dot
```

```python
# Insert new patient records
print("Adding new patient records...")
print("-" * 100)

new_records = [
    (51, "Jacob Marley", 40, "Migraine", "120/80", 70, 37.0),
    (58, "Robert Shea", 55, "Back Pain", "130/85", 75, 36.8),
    (3, "Joan Smith", 33, "Anxiety", "115/75", 68, 37.1)
]

for record in new_records:
    prms.add_patient_record(*record)
    print(f"Added: Patient ID {record[0]}, Name: {record[1]}")

print("\nAll Patient Records after insertion:")
print("-" * 100)
prms.display_all_records()
```

```python
# Visualize the tree after insertion
dot = prms.visualize_tree()
dot.render("after-insertion_tree", format="png", cleanup=True)
print("Tree after insertion saved as 'after-insertion_tree.png'")
dot
```

```python
# Display all traversals
print("=" * 100)
print("All patient records (inorder traversal):")
print("-" * 100)
for node in prms.bst.inorder_traversal(prms.bst.root):
    patient = node.value
    print(f"Patient ID: {patient.patient_id}, Name: {patient.name}, Age: {patient.age}, "
          f"Diagnosis: {patient.diagnosis}, Blood Pressure: {patient.blood_pressure}, "
          f"Pulse: {patient.pulse}, Body Temperature: {patient.body_temperature}")

print("\n" + "=" * 100)
print("All patient records (preorder traversal):")
print("-" * 100)
for node in prms.bst.preorder_traversal(prms.bst.root):
    patient = node.value
    print(f"Patient ID: {patient.patient_id}, Name: {patient.name}, Age: {patient.age}, "
          f"Diagnosis: {patient.diagnosis}, Blood Pressure: {patient.blood_pressure}, "
          f"Pulse: {patient.pulse}, Body Temperature: {patient.body_temperature}")

print("\n" + "=" * 100)
print("All patient records (postorder traversal):")
print("-" * 100)
for node in prms.bst.postorder_traversal(prms.bst.root):
    patient = node.value
    print(f"Patient ID: {patient.patient_id}, Name: {patient.name}, Age: {patient.age}, "
          f"Diagnosis: {patient.diagnosis}, Blood Pressure: {patient.blood_pressure}, "
          f"Pulse: {patient.pulse}, Body Temperature: {patient.body_temperature}")
```
